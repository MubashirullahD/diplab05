# Task 2

```python
img = cv2.imread('lena_color.jpeg', 2)
width, height = img.shape
```

```python
# To change to other types, use img.astype(np.uint8) so calculations can go negative
img = img.astype(np.float16) # BEWARE! Use float not int16.
```

```python
for y in range(height):
    # We skip the last column hence the - 1
    for x in range(1, width - 1):
        img[x][y] = (img[x + 1][y]) - (img[x][y]) 

bmax = np.amax(img)
bmin = np.amin(img)
```

```python
img = (img - bmin)/abs(bmax - bmin) * 255
img = img.astype(np.uint8)
```

```python
Image.fromarray(img, 'L')
```
